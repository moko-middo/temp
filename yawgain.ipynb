```python
import numpy as np
import matplotlib.pyplot as plt

# 
V = 120/3.6 # [m/s]
l = 2700/1000 # [m]

# A
A_list = np.arange(0.5e-3, 3.1e-3, 0.1e-3)

# 
N_list = range(12, 21)

# G
G_list = []
for N in N_list:
    G_N = []
    for A in A_list:
        G = 1 / (1 + A * V**2) * V / (N * l)
        G_N.append(G)
    G_list.append(G_N)

# 
for i in range(len(N_list)):
    plt.plot(20*np.log10(G_list[i]), A_list*1e3, label=f"N={N_list[i]}")

plt.xlim(-15, -7)
plt.ylim(0.5, 3.0)
plt.legend()
plt.xlabel("Yawgain G [dB]")
plt.ylabel("Stability Factor A [x10^(-3) s^2/m^2]")
plt.show()
```
